```python
# Removing HTML tags
import re

def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub('',text)
```

```python
# Remove Puctuation
```

```python
import string

exclude = string.punctuation
exclude = exclude + ('123456789')
def remove_punc(text):
  for char in exclude:
    if char in text:
      text = text.replace(char,'')
  return text
```

```python
df['tweet'] = df['tweet'].apply(remove_punc)
```

```python
# Converted to lower-case
chat_words = chat_words.lower()
print(chat_words)
```

```python
# converted to dictionary
abbrevation = {}

for i in chat_words.split('\n'):
  if '=' in i:
    short_full = i.split("=")
    abbrevation[short_full[0].strip()] = short_full[1].strip()
  if '–' in i:
    short_full = i.split("–")
    abbrevation[short_full[0].strip()] = short_full[1].strip()

abbrevation
```

```python
# Function for chat words treatment
# parameter -> lowercase, strip

def chat_word_treatment(text):
  new_text = []
  for w in text.split():
    if w in abbrevation:
      new_text.append(abbrevation[w])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

```python
chat_word_treatment('   pc   ')
```

```python
df['tweet'] = df['tweet'].apply(lambda x: chat_word_treatment(x))
```

```python
# Removing Stop words
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('all')
# stopwords.words('english')
```

```python
stop_words = stopwords.words('english')
def remove_stopwords(text):
  new_text = []
  for word in text.split():
    if word in stop_words:
      new_text.append('')
    else:
      new_text.append(word)
  return " ".join(new_text)
```

```python
remove_stopwords("hello my name is aravindra")
```

```python
df['tweet'] = df['tweet'].apply(remove_stopwords)
```

```python
# Stemming & #Tokenization
```

```python
#Tokenization
import spacy
nlp = spacy.load('en_core_web_sm')

# Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def token_stem_words(text):
  new_text = []

  doc1 = nlp(text)
  for word in doc1:
    new_text.append(ps.stem(word.text))

  return " ".join(new_text)

```
